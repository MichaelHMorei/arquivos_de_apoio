#Windows Function

```
# Importando as bibliotecas que serão usadas no laboratório.

from pyspark.sql import *
from pyspark.sql.types import *
from pyspark.sql.functions import *
```

```
path = '/FileStore/tables/titanic.csv'
file_type = 'csv'

df_titanic = spark \
            .read.format(file_type) \
            .option("inferSchema", "True") \
            .option("sep", ",") \
            .option("header", "True") \
            .csv(path)
            
```

```
df_titanic.show(10)
```

```
df_titanic.withColumnRenamed("Name", "Nome").show(5)
```

```
#Criando Coluna
df_titanic.withColumn("Nova_Coluna", lit('na')).show(5)
#Criando função para preencher a Coluna
repla = udf(lambda x: 'M' if x == 'male'
                       else 'F')
```

```
#Chamando a função para preencher a coluna, passando a coluna [Sex] como parâmetro.
df_titanic.withColumn('Nova_Coluna', repla('Sex')).show(5)
```

```
#Coando where funciona igual ao comando Filter
df_titanic.where(col("Survived") == 1).show()
```

```
#Comando whare com a condicional & - and
df_titanic.where((df_titanic['Survived'] == 1) & (df_titanic["Sex"] == 'female')).show(5)
```

```
#Descrevendo os dados , função estatistica
df_titanic.describe().show()
```
